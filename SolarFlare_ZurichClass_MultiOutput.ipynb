```python
import pandas as pd

flare_data = pd.read_csv('/content/flare.data1', sep=' ', header=0)
flare_data
```

```python
flare_data = pd.get_dummies(flare_data, columns=['Y1', 'Y2', 'Y3'])
flare_data
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input
from sklearn.model_selection import train_test_split
```

```python
def format_output(data):
  y_B = data.pop('Y1_B')
  y_B = np.array(y_B)
  y_C = data.pop('Y1_C')
  y_C = np.array(y_C)
  y_D = data.pop('Y1_D')
  y_D = np.array(y_D)
  y_E = data.pop('Y1_E')
  y_E = np.array(y_E)
  y_F = data.pop('Y1_F')
  y_F = np.array(y_F)
  y_H = data.pop('Y1_H')
  y_H = np.array(y_H)
  return y_B, y_C, y_D, y_E, y_F, y_H

def norm(x):
  return (x - train_stats['mean']) / train_stats['std']

def plot_diff(y_true, y_pred, title=''):
  plt.scatter(y_true, y_pred)
  plt.title(title)
  plt.xlabel('True Values')
  plt.ylabel('Predictions')
  plt.axis('equal')
  plt.axis('square')
  plt.xlim(plt.xlim())
  plt.ylim(plt.ylim())
  plt.plot([-100, 100], [-100, 100])
  plt.show()

def plot_metrics(metric_name, title, ylim=5):
  plt.title(title)
  plt.ylim(0, ylim)
  plt.plot(history.history[metric_name], color='blue', label=metric_name)
  plt.plot(history.history['val_' + metric_name], color='green', label='val_' + metric_name)
```

```python
train, test = train_test_split(flare_data, test_size=0.2)
train_stats = train.describe()

train_stats.pop('Y1_B')
train_stats.pop('Y1_C')
train_stats.pop('Y1_D')
train_stats.pop('Y1_E')
train_stats.pop('Y1_F')
train_stats.pop('Y1_H')
train_stats = train_stats.transpose()
train_y = format_output(train)
test_y = format_output(test)

norm_train_X = norm(train)
norm_test_X = norm(test)
```

```python
input_layer = Input(shape=(len(train .columns), ))
first_dense = Dense(128, activation='relu')(input_layer)
second_dense = Dense(128, activation='relu')(first_dense)

third_dense = Dense(64, activation='relu')(second_dense)
yB_output = Dense(6, name='Y1_B_output')(third_dense)

fourth_dense = Dense(64, activation='relu')(third_dense)
yC_output = Dense(6, name='Y1_C_output')(fourth_dense)

fifth_dense = Dense(64, activation='relu')(fourth_dense)
yD_output = Dense(6, name='Y1_D_output')(fifth_dense)

sixth_dense = Dense(64, activation='relu')(fifth_dense)
yE_output = Dense(6, name='Y1_E_output')(sixth_dense)

seventh_dense = Dense(64, activation='relu')(sixth_dense)
yF_output = Dense(6, name='Y1_F_output')(seventh_dense)

eighth_dense = Dense(64, activation='relu')(seventh_dense)
yH_output = Dense(6, name='Y1_H_output')(eighth_dense)

model = Model(inputs=input_layer, outputs=[yB_output, yC_output, yD_output, yE_output, yF_output, yH_output])
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss = {'Y1_B_output': 'binary_crossentropy',
                      'Y1_C_output': 'binary_crossentropy',
                      'Y1_D_output': 'binary_crossentropy',
                      'Y1_E_output': 'binary_crossentropy',
                      'Y1_F_output': 'binary_crossentropy',
                      'Y1_H_output': 'binary_crossentropy'},
              metrics = {'Y1_B_output': tf.keras.metrics.BinaryAccuracy(),
                         'Y1_C_output': tf.keras.metrics.BinaryAccuracy(),
                         'Y1_D_output': tf.keras.metrics.BinaryAccuracy(),
                         'Y1_E_output': tf.keras.metrics.BinaryAccuracy(),
                         'Y1_F_output': tf.keras.metrics.BinaryAccuracy(),
                         'Y1_H_output': tf.keras.metrics.BinaryAccuracy()})
```
